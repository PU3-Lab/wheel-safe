```python
import cv2
import numpy as np

from lib.utils.path import raw_data_path
from v3.slope_pipeline import COLOR_MAP, SlopePipeline
from v3.viz_map import report_2, report_road

image_names = ['ZED1_KSC_042960']

foder_nums = ['024']

slope_pipeline = SlopePipeline()

for _, (name, num) in enumerate(list(zip(image_names, foder_nums, strict=False))):
    folder_path = raw_data_path(num)
    conf_file = folder_path / f'Depth_{num}.conf'
    img_path = folder_path / f'{name}_left.png'
    conf_img = folder_path / f'{name}_confidence.png'
    disp_img = folder_path / f'{name}_disp16.png'

    disp_map = cv2.imread(str(disp_img), cv2.IMREAD_UNCHANGED).astype(np.float32) / 16
    conf_map = (
        cv2.imread(str(conf_img), cv2.IMREAD_UNCHANGED).astype(np.float32) / 255.0
    )

    slope_pipeline.set_config_params(conf_file)

    pred, img = slope_pipeline.run(img_path)
    slope, mask = slope_pipeline.estimate(conf_map, disp_map)

    palette = np.array(COLOR_MAP, dtype=np.uint8)
    report_2(img_path, pred, conf_map, disp_map, palette)

    if slope is not None:
        report_road(img_path, mask, slope)
```
